```python
# Add dummy variables 
df2['negemo higher than average'] = np.where(df2['negemo']>1.316642314, 1, 0)
df2['anx higher than average'] = np.where(df2['anx']>0.224144827978823, 1, 0)
df2['anger higher than average'] = np.where(df2['anger']>0.359500920355999, 1, 0)
df2['sad higher than average'] = np.where(df2['sad']>0.275241474846767, 1, 0)
df2.head()
```

```python
#drop orginal columns
df3= df2.drop(columns=["negemo","anx","anger", "sad"])
df3.head()
```

```python
# Dummy-encode the categorical features

CATEGORICAL_FEATURES = ["anx higher than average", "anger higher than average", "sad higher than average"]
encoded_df = pd.get_dummies(df3, columns=CATEGORICAL_FEATURES)
encoded_df
```

```python
# Standard-scale continuous numeric features

from sklearn.preprocessing import StandardScaler

# Standard-scale only continuous numeric features
CONTINUOUS_NUMERIC_FEATURES = ["posemo", "score", "pronoun", "article","auxverb", "adverb", "conj", "verb", "adj","cogproc", "focuspast", "focuspresent", "focusfuture", "work", "informal"]
scaled_df = encoded_df.copy()
scaled_df[CONTINUOUS_NUMERIC_FEATURES] = StandardScaler().fit_transform(scaled_df[CONTINUOUS_NUMERIC_FEATURES])
scaled_df
```

```python
#Import Module
from sklearn.model_selection import train_test_split
X=encoded_df.drop(['negemo higher than average'],axis=1)
y=encoded_df['negemo higher than average']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=52697)
```

```python
print('All:', np.bincount(y) / float(len(y)) * 100.0)
print('Training:', np.bincount(y_train) / float(len(y_train)) * 100.0)
print('Test:', np.bincount(y_test) / float(len(y_test)) * 100.0)
```

```python
#LogisticRegression with accuracy
from sklearn import metrics
from sklearn.linear_model import LogisticRegression

dt = LogisticRegression(random_state=0, solver='newton-cg')
dt.fit(X_train, y_train)
y_train_pred_b = dt.predict(X_train)
y_test_pred_b = dt.predict(X_test)

accuracy_train=metrics.accuracy_score(y_train, y_train_pred_b)
accuracy_test=metrics.accuracy_score(y_test,y_test_pred_b)
print(y_train_pred_b.sum(),y_test_pred_b.sum(),round(accuracy_train,2),round(accuracy_test,2 ))
```

```python
#It is often helpful to compare the generalization performance (on the test set) to the performance on the training set:
dt.score(X_train, y_train)
```

```python
#DecisionTreeRegressor with accuracy
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics

dt = DecisionTreeRegressor(max_depth=9, random_state=0)
dt.fit(X_train, y_train)
y_train_pred = dt.predict(X_train)
y_test_pred = dt.predict(X_test)

r2_train=metrics.r2_score(y_train, y_train_pred)
r2_test=metrics.r2_score(y_test, y_test_pred)
print(y_train_pred.sum(),y_test_pred.sum(),round(r2_train,2),round(r2_test,2 ))
```

```python
from sklearn.neighbors import KNeighborsClassifier
#This creates a model object.
classifier = KNeighborsClassifier()
#This fits the model object to the data.
classifier.fit(X_train, y_train)
#This creates the prediction. 
pred_y = classifier.predict(X_test)
```

```python
from sklearn import metrics

#This calculates the accuracy.
print("Manual Accuracy:", np.sum(pred_y == y_test) / float(len(y_test)))
print("Metrics score: ", metrics.accuracy_score(y_test, pred_y) )
print("Classifier score: ", classifier.score(X_test, y_test) )
```
